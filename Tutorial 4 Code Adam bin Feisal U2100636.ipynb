```python
from scipy.fftpack import fft, ifft
import matplotlib.pyplot as plt
import numpy as np
def F1(t):
    return 2*np.sin(2.0*np.pi*t)
def F2(t):
    return 2*np.sin(2.0*np.pi*t)
def F3(t):
    return 2*np.sin(2.0*np.pi*t)
def F(t):
    return F1(t)+F2(t)+F3(t)

#number of sample points
N = 800
#sample spacing
T = 1.0/N

t=np.linspace(0.0,N*T,N)

#original function
y=F(t)

#Original Function plotting
plt.figure(1)
plt.title('$f(x)=2sin (2 \pi t) + 2sin(2 \pi t) + 2sin (2 \pi t) $')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.plot(t, y)
plt.legend()
plt.grid()
```

```python
#perform dft by calling the scipy library
yf = fft(y)
tf = np.linspace(0.0, 1.0/(2.0*T), N//2)

#Fourier Transform plotting
plt.figure(2)
plt.title('Fourier Transform of F(x)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.plot(tf,  2.0/N * np.abs(yf[0:N//2]))
plt.grid()
```

```python
# Inverse Fourier Transform
yinv = ifft(yf)

# Plot Inverse Fourier Transform
plt.figure(3)
plt.plot(t, yinv)
plt.title('Inverse Fourier Transform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
```
