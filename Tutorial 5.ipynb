The Analytical Value comes from direct integration of the equation (where possible).

The Trapezoid Rule and Simpson's (D'oh!) Rule values are given from scipy integration functions.

Question 1(a)

<ins> Analytical Value </ins>

$ \frac {e^{-x}sin(x)}{2} - \frac {e^{-x}cos(x)}{2} +C $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def f(x):
    return np.exp(-x)*np.cos(x)

x = np.linspace(0,1,100)
y = f(x)

#Analytical solution
def integral(x):
    return (np.exp(-x)*np.sin(x)/2)-(np.exp(-x)*np.cos(x)/2)

def area_integral(a,b):
    return integral(b)-integral(a)
area_direct = area_integral(0,1)

#Simpson's Rule
area_simps = integrate.simpson(y,x=x)

#Trapezoid rule
area_trapezoid = integrate.trapezoid(y,x)

#Monte Carlo
width = 1-0
samples= np.random.uniform(low=0,high=width,size=100000) #random number generator, the last number means the amount of numbers generated
area_mc = f(samples).mean()*width 
error_mc=np.std(samples*width) / np.sqrt(samples.size)



#Graph plotting
plt.plot(x,y)
plt.fill_between(x,0,y,alpha=0.1)
plt.text(0.05,0.3,f"area from direct integration is {area_direct:0.3f}",fontsize=12)
plt.text(0.05,0.2,f"area from Simpson Rule is {area_simps:0.3f}",fontsize=12)
plt.text(0.05,0.1,f"area from Trapezoid Rule is {area_trapezoid:0.3f}",fontsize=12)
plt.text(0.05,0.0,f"area from Monte Carlo Integration is {area_mc:0.3f}{error_mc:0.3f}",fontsize=12)
plt.xlabel('x')
plt.ylabel('f(x)')
```

This integral can only be solved analytically using Series expansion.







```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

def f(x):
    return np.exp(np.sin(2.0*x))

x = np.linspace(0,2.0*np.pi,100)
y = f(x)

#Simpson's Rule
area_simps = simps(y,x=x)

#Trapezoid rule
area_trapezoid = integrate.trapezoid(y,x)

#Monte Carlo
width = 2.0*np.pi-0
samples= np.random.uniform(low=0,high=width,size=100000) #random number generator, the last number means the amount of numbers generated
area_mc = f(samples).mean()*width 
error_mc=np.std(samples*width) / np.sqrt(samples.size)

#Graph plotting
plt.plot(x,y)
plt.fill_between(x,0,y,alpha=0.1)
plt.text(0.1,0.2,f"area from Simpson Rule is {area_simps:0.3f}",fontsize=8)
plt.text(0.1,0.1,f"area from Trapezoid Rule is {area_trapezoid:0.3f}",fontsize=8)
plt.text(0.1,0.0,f"area from Monte Carlo Integration is {area_mc:0.3f}{error_mc:0.3f}",fontsize=8)
plt.xlabel('x')
plt.ylabel('f(x)')
```

Question 1(c)

<ins> Analytical Value </ins>

$ \frac {2}{\sqrt 3} tan^{-1} (\frac {1}{\sqrt 3} tan(\frac {x}{2})) + C $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

def f(x):
    return 1/(2+np.cos(x))

x = np.linspace(0,2.0*np.pi,100)
y = f(x)

#Analytical solution
def integral(x):
    return 2.0*np.arctan(1.0*np.tan(x/2.0)/np.sqrt(3.0))/np.sqrt(3.0)

def area_integral(a,b):
    return integral(b)-integral(a)
area_direct = area_integral(0,np.pi)*2

#Simpson's Rule
area_simps = simps(y,x=x)

#Trapezoid rule
area_trapezoid = integrate.trapezoid(y,x)

#Monte Carlo
width = 2.0*np.pi-0
samples= np.random.uniform(low=0,high=width,size=100000) #random number generator, the last number means the amount of numbers generated
area_mc = f(samples).mean()*width 
error_mc=np.std(samples*width) / np.sqrt(samples.size)

#Graph plotting
plt.plot(x,y)
plt.fill_between(x,0,y,alpha=0.1)
plt.text(0.05,0.3,f"area from direct integration is {area_direct:0.3f}",fontsize=8)
plt.text(0.1,0.2,f"area from Simpson Rule is {area_simps:0.3f}",fontsize=8)
plt.text(0.1,0.1,f"area from Trapezoid Rule is {area_trapezoid:0.3f}",fontsize=8)
plt.text(0.1,0.0,f"area from Monte Carlo Integration is {area_mc:0.3f}{error_mc:0.3f}",fontsize=8)
plt.xlabel('x')
plt.ylabel('f(x)')
```

Question 2

For this question, we have to find the normalisation constant, a value that when multiplied to the original integral, will give a value of 1. This is found by taking the reciprocal of the original integral.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

def f(x):
    return np.exp(-x)/x

x = np.linspace(1e-6,1,100)
y = f(x)

#Simpson's Rule
area_simps = simps(y,x=x)

#Trapezoid rule
area_trapezoid = integrate.trapezoid(y,x)

#Monte Carlo
width = 1-1e-6
samples= np.random.uniform(low=0,high=width,size=10000000) #random number generator, the last number means the amount of numbers generated
area_mc = f(samples).mean()*width 
error_mc=np.std(samples*width) / np.sqrt(samples.size)

#Graph plotting
plt.figure(1)
plt.title('Graph of function')
plt.plot(x,y)
plt.text(0.05,100000,f"Normalistion constant from Monte Carlo Integration is {area_mc:0.4f}{error_mc:0.4f}",fontsize=8)
plt.text(0.05,200000,f"Normalistion constant  from Simpson Rule is {area_simps:0.3f}",fontsize=8)
plt.text(0.05,300000,f"Normalistion constant  from Trapezoid Rule is {area_trapezoid:0.3f}",fontsize=8)
plt.xlabel('x')
plt.ylabel('f(x)')

#Normalised function
plt.figure(2)
plt.title('Graph of normalised function')
plt.plot(x,y*(1/area_mc), label='Normalisation constant from Monte Carlo')
plt.plot(x,y*(1/area_simps), label='Normalisation constant from Simpsons rule')
plt.plot(x,y*(1/area_trapezoid), label='Normalisation constant from trapezoidal rule')
plt.legend(loc='best')
plt.ylim(0,0.1)
plt.xlabel('x')
plt.ylabel('f(x)')
```
